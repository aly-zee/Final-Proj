```python
# drop ~350 rows
retail.dropna(inplace=True)
retail.isnull().sum()
```

```python
# view unique variables in the Product Type column
retail["Product_Category"].unique()
```

```python
# convert Total_purchase column from float to int
cols_to_int = ["Total_Purchases","Income"]

for col in cols_to_int:
    retail[col] = retail[col].astype(int)
```

```python
# format Total_amount, column to be two decimal places
retail["Total_Amount"] = retail["Total_Amount"].map('{:.2f}'.format)
retail["Total_Amount"]
```

```python
# convert columns from int to str objects
cols_to_objects = ["Transaction_ID","Customer_ID","Phone","Zipcode","Age"]

for cols in cols_to_objects:
    retail[cols] = retail[cols].astype(str)
    retail[cols] = retail[cols].apply(lambda x: x.split('.')[0])
```

## Bivariate Visualization

```python
# product type purchases by gender (Categorical Vs. Numerical)
# crosstab 
crosstab_gender_purchases = pd.crosstab(retail["Product_Category"],retail["Gender"])
crosstab_gender_purchases
```

```python

chi2_statistic, chi2_p_value, _, _ = chi2_contingency(crosstab_gender_purchases)

chi2_statistic, chi2_p_value
if chi2_p_value < 0.05:
    print("Reject null hypothesis. Purchases by gender have a significant association")
else:
    print("Dont reject null")
```

```python
# plot the cross tab
crosstab_gender_purchases.plot(kind="bar", stacked = True);
```

# Unsupervised Learning: K-Means Clustering

```python
frequency_DF = retail.groupby(["Phone"]).agg({
    'Order_Status':'count'
})
```

```python
frequency_DF[frequency_DF["Order_Status"]>2]
```

```python
'''
Features for the k-means clustering:
1. Gender
2. Monetary Value
2. Income
3. Age
4. Product Type - Type of product purchased
5. Customer Loyalty level
''' 
columns_to_drop = ["Transaction_ID","Customer_ID","Name","Email","Phone","Address","City","State", "Zipcode","Country","Date","Year","Month","Time","Total_Purchases","Amount","Product_Brand","Product_Type","Feedback","Shipping_Method","Order_Status","Ratings","products","Payment_Method","Gender","Product_Category","Customer_Segment"]

features = retail.drop(columns_to_drop, axis=1)
```

```python
scaler_minmax = MinMaxScaler()
features_scaled_x = scaler_minmax.fit_transform(features)
features_scaled = pd.DataFrame(features_scaled_x,columns = features.columns)
```

```python
features_scaled
```

```python
inertia =  []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42,n_init=10)
    kmeans.fit(features_scaled)
    inertia.append(kmeans.inertia_)
    
# plot kmeans
plt.plot(range(1,11), inertia,marker="o")
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('KMeans Clusters')
plt.show()
```

```python
km = KMeans(n_clusters = 4).fit(features_scaled)
retail["Cluster"] = km.labels_
```

```python
retail[retail["Cluster"]==1]
```

```python
# standard scaler
scaler = StandardScaler()
```

```python
scaled_features = features_scaled = pd.DataFrame(scaler.fit_transform(features), columns=['Age', 'Income', 'Amount Spent'])
```

```python
k=4
kmeans = KMeans(n_clusters=k,random_state=42)
kmeans.fit(scaled_features)
```

```python
# Get cluster labels and centroids
cluster_labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_
```

```python
retail['Cluster'] = kmeans.labels_
```
